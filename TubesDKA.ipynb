```python
# MODUL: INISIALISASI ARMADA (TRUK & GEROBAK)
class Kendaraan:
    def __init__(self, id_kendaraan, tipe, kapasitas_maks, waktu_mulai, waktu_selesai, posisi_awal):
        self.id = id_kendaraan
        self.tipe = tipe # "Gerobak" atau "Truk"
        self.kapasitas_maks = kapasitas_maks
        self.muatan = 0
        self.waktu_aktif = waktu_mulai # Disimpan dalam menit (misal 06:00 = 360)
        self.waktu_selesai = waktu_selesai
        self.jarak_tempuh = 0
        self.posisi = posisi_awal
        self.log = []

    def catat_log(self, pesan):
        # Mengubah format menit kembali ke format Jam:Menit untuk pelaporan
        jam = self.waktu_aktif // 60
        menit = self.waktu_aktif % 60
        self.log.append(f"[{jam:02d}:{menit:02d}] {self.id}: {pesan}")

# 1. SETUP ARMADA TRUK
# Aturan: Kapasitas 200kg, Operasi 08:00 (480) - 17:00 (1020)
# Start: Truk 1 di TPS_1, Truk 2 di TPS_2, Truk 3 di TPS_3
armada_truk = [
    Kendaraan("Truk_1", "Truk", 200, 480, 1020, "TPS_1"),
    Kendaraan("Truk_2", "Truk", 200, 480, 1020, "TPS_2"),
    Kendaraan("Truk_3", "Truk", 200, 480, 1020, "TPS_3")
]

# 2. SETUP ARMADA GEROBAK
# Aturan: Kapasitas 15kg, Operasi 06:00 (360) - 15:00 (900)
# Start: 2 di TPS_1, 3 di TPS_2, 2 di TPS_3
armada_gerobak = [
    # 2 Gerobak start di TPS_1
    Kendaraan("Gerobak_1", "Gerobak", 15, 360, 900, "TPS_1"),
    Kendaraan("Gerobak_2", "Gerobak", 15, 360, 900, "TPS_1"),

    # 3 Gerobak start di TPS_2
    Kendaraan("Gerobak_3", "Gerobak", 15, 360, 900, "TPS_2"),
    Kendaraan("Gerobak_4", "Gerobak", 15, 360, 900, "TPS_2"),
    Kendaraan("Gerobak_5", "Gerobak", 15, 360, 900, "TPS_2"),

    # 2 Gerobak start di TPS_3
    Kendaraan("Gerobak_6", "Gerobak", 15, 360, 900, "TPS_3"),
    Kendaraan("Gerobak_7", "Gerobak", 15, 360, 900, "TPS_3")
]

# Output Pengecekan
print("--- STATUS ARMADA SIAP OPERASI ---")
print(f"[TRUK]    : {len(armada_truk)} unit terdaftar (Start di TPS 1, 2, dan 3).")
print(f"[GEROBAK] : {len(armada_gerobak)} unit terdaftar (Distribusi start: 2 unit di TPS_1, 3 unit di TPS_2, 2 unit di TPS_3).")
print("Semua armada menunggu instruksi pergerakan (Pathfinding).")
```

```python
import networkx as nx

# MODUL: PENCARIAN RUTE & GPS ARMADA (PATHFINDING CORE)

# 1. MEMBUAT PETA KHUSUS TRUK (Membuang jalan gang)
# Truk hanya boleh lewat jalan yang aksesnya "Semua"
edges_truk = [(u, v) for u, v, d in G.edges(data=True) if d['akses'] == "Semua"]
G_truk = G.edge_subgraph(edges_truk).copy()

def cari_target_terdekat(posisi_awal, kandidat_target, tipe_kendaraan):
    """
    Fungsi GPS (Dijkstra Algorithm):
    Mencari target dengan jarak (weight) terpendek dari posisi awal.
    """
    # Tentukan peta yang dipakai: Truk pakai G_truk, Gerobak pakai G penuh
    peta_aktif = G_truk if tipe_kendaraan == "Truk" else G

    target_terpilih = None
    jarak_terpendek = float('inf')

    # Keamanan: Pastikan posisi awal ada di peta kendaraan tersebut
    if posisi_awal not in peta_aktif:
        return None, float('inf')

    # Bandingkan jarak ke setiap kandidat target
    for target in kandidat_target:
        # Lewati jika target kebetulan tidak ada di peta kendaraan (misal truk mencari rumah gang)
        if target not in peta_aktif:
            continue

        try:
            # Mencari jarak rute terpendek dengan menjumlahkan 'weight' di setiap ruas jalan
            jarak = nx.shortest_path_length(peta_aktif, source=posisi_awal, target=target, weight='weight')

            # Jika ditemukan jarak yang lebih dekat dari yang tersimpan, jadikan ini target utama
            if jarak < jarak_terpendek:
                jarak_terpendek = jarak
                target_terpilih = target

        except nx.NetworkXNoPath:
            # Lewati jika kebetulan tidak ada jalan menuju target tersebut
            continue

    return target_terpilih, jarak_terpendek

# 2. FUNGSI PEMBANTU (HELPER) UNTUK MENDAPATKAN KANDIDAT
def dapatkan_rumah_kotor(graf):
    """Mengambil list ID rumah yang masih memiliki sampah > 0"""
    return [n for n, d in graf.nodes(data=True) if d.get('tipe') == 'Rumah' and d.get('sampah') > 0]

def dapatkan_semua_tps(graf, muatan_kendaraan=0):
    """Mengambil list ID TPS yang masih sanggup menampung muatan armada saat ini"""
    return [n for n, d in graf.nodes(data=True) if d.get('tipe') == 'TPS' and d.get('kapasitas_sisa', 0) >= muatan_kendaraan]

# Output Pengecekan
print("--- STATUS MODUL PENCARIAN RUTE ---")
print(f"Peta Utama (Gerobak) : {G.number_of_edges()} ruas jalan terdeteksi.")
print(f"Peta Khusus (Truk)   : {G_truk.number_of_edges()} ruas jalan besar terdeteksi (Jalan gang di-filter).")
print("Sistem GPS / Pathfinding berbasis Dijkstra siap digunakan.")

# Contoh Uji Coba GPS Kecil-kecilan (Bisa dihapus nanti, hanya untuk tes)
# Kita tes mencari TPS terdekat dari rumah RU_42 menggunakan Gerobak
contoh_tps, jarak_contoh = cari_target_terdekat("RU_42", dapatkan_semua_tps(G), "Gerobak")
print(f"\n[Test GPS] TPS terdekat dari RU_42 adalah {contoh_tps} dengan jarak {jarak_contoh} satuan.")
```

```python
import math

# MODUL: ENGINE SIMULASI KRONOLOGIS (THE ENGINE)
print("Memulai Simulasi Pengangkutan Sampah Desa (06:00 - 17:00)...")

# Gabungkan seluruh armada untuk diproses secara bersamaan
semua_armada = armada_gerobak + armada_truk
simulasi_selesai = False

while not simulasi_selesai:
    # 1. AMBIL KENDARAAN YANG GILIRANNYA PALING AWAL
    armada_aktif = [k for k in semua_armada if k.waktu_aktif < k.waktu_selesai]

    if not armada_aktif:
        simulasi_selesai = True
        break

    armada_aktif.sort(key=lambda x: x.waktu_aktif)
    k = armada_aktif[0]

    rumah_kotor = dapatkan_rumah_kotor(G)

    # SKENARIO A: KENDARAAN PENUH (Atau desa bersih tapi bawa sisa sampah)
    if k.muatan >= k.kapasitas_maks or (not rumah_kotor and k.muatan > 0):
        target_buang = None
        is_tps = True
        jarak_tempuh = float('inf')

        # [PERBAIKAN 1]: Cek TPS yang sisa kapasitasnya masih sanggup menampung!
        tps_tersedia = dapatkan_semua_tps(G, k.muatan)
        if not tps_tersedia:
            k.catat_log("Semua TPS Penuh! Operasi dibatalkan karena tidak ada TPS yang muat.")
            k.waktu_aktif = k.waktu_selesai
            continue

        tps_terdekat, jarak_tps = cari_target_terdekat(k.posisi, tps_tersedia, k.tipe)

        if k.tipe == "Gerobak":
            truk_tersedia = [t for t in armada_truk if (t.kapasitas_maks - t.muatan) >= k.muatan and t.waktu_aktif <= k.waktu_aktif]
            posisi_truk_tersedia = [t.posisi for t in truk_tersedia]

            truk_terdekat, jarak_truk = cari_target_terdekat(k.posisi, posisi_truk_tersedia, "Semua")

            if truk_terdekat and jarak_truk < jarak_tps:
                target_buang = truk_terdekat
                jarak_tempuh = jarak_truk
                is_tps = False

                # [PERBAIKAN 2]: Hentikan/tambah waktu aktif Truk saat dititipi sampah
                for t in truk_tersedia:
                    if t.posisi == target_buang:
                        t.muatan += k.muatan
                        waktu_tunggu_truk = k.muatan * 2
                        t.waktu_aktif += waktu_tunggu_truk # Truk pause bergerak selama bongkar muat
                        t.catat_log(f"Menerima titipan {k.muatan}kg dari {k.id}. Berhenti loading {waktu_tunggu_truk} menit.")
                        break
            else:
                target_buang = tps_terdekat
                jarak_tempuh = jarak_tps
        else:
            target_buang = tps_terdekat
            jarak_tempuh = jarak_tps

        if target_buang is None:
            k.waktu_aktif += 1
            continue

        # PENGHITUNGAN WAKTU BONGKAR MUAT
        if k.tipe == "Gerobak":
            waktu_jalan = jarak_tempuh * 3
            waktu_load = k.muatan * 2
        else:
            waktu_jalan = math.ceil(jarak_tempuh * 0.6)
            waktu_load = math.ceil((k.muatan / 10) * 2)

        # [PERBAIKAN 3]: Validasi Lembur sebelum mengeksekusi keberangkatan
        if (k.waktu_aktif + waktu_jalan + waktu_load) > k.waktu_selesai:
            k.catat_log("Waktu operasi tidak cukup untuk ke target. Batal berangkat dan kembali ke pool.")
            k.waktu_aktif = k.waktu_selesai
            continue

        # Update Status Kendaraan
        k.waktu_aktif += (waktu_jalan + waktu_load)
        k.jarak_tempuh += jarak_tempuh

        if is_tps:
            # [PERBAIKAN 1]: Kurangi sisa kapasitas TPS secara matematis
            G.nodes[target_buang]['kapasitas_sisa'] -= k.muatan
            G.nodes[target_buang]['terisi'] += k.muatan # (Tetap dijumlahkan untuk keperluan Laporan Poin F)
            k.catat_log(f"Buang {k.muatan}kg ke {target_buang} (Jarak: {jarak_tempuh} | Sisa Kuota TPS: {G.nodes[target_buang]['kapasitas_sisa']}kg)")
        else:
            k.catat_log(f"Berpapasan! Titip {k.muatan}kg ke Truk di {target_buang} (Jarak: {jarak_tempuh})")

        k.posisi = target_buang
        k.muatan = 0

    # SKENARIO B: ADA RUANG DI MUATAN & MASIH ADA SAMPAH DI DESA
    elif rumah_kotor:
        target, jarak = cari_target_terdekat(k.posisi, rumah_kotor, k.tipe)

        if not target:
            # [PERBAIKAN FINAL]: Jangan paksa selesai! Biarkan kendaraan menunggu 15 menit (idle)
            # agar tetap eksis di garis waktu dan bisa dititipi sampah oleh gerobak.
            k.waktu_aktif += 15
            k.catat_log("Tidak ada akses jalan. Menunggu 15 menit barangkali ada gerobak melintas...")
            continue

        sampah_tersedia = G.nodes[target]['sampah']
        sisa_kapasitas = k.kapasitas_maks - k.muatan
        sampah_diambil = min(sampah_tersedia, sisa_kapasitas)

        if k.tipe == "Gerobak":
            waktu_jalan = jarak * 3
        else:
            waktu_jalan = math.ceil(jarak * 0.6)

        waktu_load = sampah_diambil * 2

        # [PERBAIKAN 3]: Validasi Lembur sebelum mengeksekusi keberangkatan
        if (k.waktu_aktif + waktu_jalan + waktu_load) > k.waktu_selesai:
            k.catat_log("Waktu operasi tidak cukup untuk mengambil sampah target. Selesai hari ini.")
            k.waktu_aktif = k.waktu_selesai
            continue

        # Update Status Kendaraan
        k.waktu_aktif += (waktu_jalan + waktu_load)
        k.jarak_tempuh += jarak
        k.muatan += sampah_diambil
        G.nodes[target]['sampah'] -= sampah_diambil
        k.posisi = target

        k.catat_log(f"Ambil {sampah_diambil}kg dari {target} (Muatan: {k.muatan}/{k.kapasitas_maks}kg)")

    # SKENARIO C: DESA BERSIH
    else:
        k.waktu_aktif = k.waktu_selesai
        k.catat_log("Desa sudah bersih! Selesai operasi hari ini.")

print("Engine Stop: Seluruh armada telah menyelesaikan jam operasinya atau desa sudah bersih.")
```

```python
# MODUL 3: PELAPORAN AKHIR & TRANSPARANSI TUGAS AKHIR

print("="*70)
print("LAPORAN AKHIR PENGELOLAAN SAMPAH DESA - PAK KADES")
print("="*70)

# 1. LAPORAN JARAK & WAKTU GEROBAK
print("\n[A] KINERJA ARMADA GEROBAK (Jam Operasional: 06:00 - 15:00)")
total_jarak_gerobak = 0
for g in armada_gerobak:
    # Menghitung total waktu kerja dari jam 06:00 (menit 360)
    durasi_menit = g.waktu_aktif - 360
    jam = max(0, durasi_menit // 60)
    menit = max(0, durasi_menit % 60)

    print(f" - {g.id:10} | Jarak: {g.jarak_tempuh:3} satuan | Waktu Kerja: {jam} jam {menit} menit")
    total_jarak_gerobak += g.jarak_tempuh

print(f" >> TOTAL JARAK KESELURUHAN GEROBAK: {total_jarak_gerobak} satuan")


# 2. LAPORAN JARAK & WAKTU TRUK
print("\n[B] KINERJA ARMADA TRUK (Jam Operasional: 08:00 - 17:00)")
total_jarak_truk = 0
for t in armada_truk:
    # Menghitung total waktu kerja dari jam 08:00 (menit 480)
    durasi_menit = t.waktu_aktif - 480
    jam = max(0, durasi_menit // 60)
    menit = max(0, durasi_menit % 60)

    print(f" - {t.id:10} | Jarak: {t.jarak_tempuh:3} satuan | Waktu Kerja: {jam} jam {menit} menit")
    total_jarak_truk += t.jarak_tempuh

print(f" >> TOTAL JARAK KESELURUHAN TRUK: {total_jarak_truk} satuan")


# 3. LAPORAN SISA SAMPAH (POIN F)
print("\n[C] STATUS AKHIR SAMPAH DESA (Pukul 17:00)")
sisa_di_rumah = sum([d.get('sampah') for n, d in G.nodes(data=True) if d.get('tipe') == 'Rumah'])
sisa_di_gerobak = sum([g.muatan for g in armada_gerobak])
sisa_di_truk = sum([t.muatan for t in armada_truk])
sampah_di_tps = sum([d.get('terisi') for n, d in G.nodes(data=True) if d.get('tipe') == 'TPS'])

print(f" 1. Sampah Warga yang belum terangkut (di rumah) : {sisa_di_rumah} kg")
print(f" 2. Sampah tertahan di bak Gerobak               : {sisa_di_gerobak} kg")
print(f" 3. Sampah tertahan di bak Truk                  : {sisa_di_truk} kg")
print(f" 4. TOTAL SAMPAH BERHASIL DIOLAH DI TPS          : {sampah_di_tps} kg")


# 4. BUKTI TRANSPARANSI (LOG PERGERAKAN)
print("\n[D] LOG TRANSPARANSI PERGERAKAN ARMADA (Buku Catatan)")
for k in semua_armada:
    print(f"\n>> {k.id} (Tugas selesai di posisi: {k.posisi}):")
    if not k.log:
        print("   [!] Tidak ada pergerakan / Menunggu instruksi.")
    else:
        for aktivitas in k.log:
            print(f"   {aktivitas}")

print("\n" + "="*70)
print("SIMULASI SELESAI. LAPORAN SIAP DISERAHKAN KE DOSEN PENGUJI.")
print("="*70)
```
