```python
import random

# ==============================================================================
# INISIALISASI VARIABEL GLOBAL (STATE & EVENT TRACKER)
# ==============================================================================
house_waste = {}        # Melacak tumpukan sampah di depan rumah saat ini
total_initial_waste = 0 # Akumulasi total sampah satu desa hari ini
waste_events = {}       # Jadwal event warga membuang sampah {menit: [(node, kg)]}

# ==============================================================================
# MODUL: GENERATOR EVENT SAMPAH DINAMIS (BERBASIS PROBABILITAS STATISTIK)
# ==============================================================================
def generate_jadwal_sampah(graf):
    global total_initial_waste, house_waste, waste_events
    total_sampah = 0
    jumlah_rumah = 0
    
    house_waste = {}
    waste_events = {}

    # Konfigurasi Probabilitas Berat Sampah Harian (Sangat Realistis)
    # 1-3 kg (Paling umum), 4-5 kg (Jarang), 6-7 kg (Sangat langka), 0 kg (Rumah kosong/Libur)
    berat_opsi = [0, 1, 2, 3, 4, 5, 6, 7]
    berat_bobot = [0.05, 0.30, 0.35, 0.15, 0.08, 0.04, 0.02, 0.01] 
    
    # Konfigurasi Probabilitas Frekuensi Buang Sampah
    freq_opsi = [1, 2, 3]
    freq_bobot = [0.60, 0.35, 0.05] # 60% buang sekali, 35% dua kali, 5% tiga kali

    for node, data in graf.nodes(data=True):
        if data.get('tipe') == 'Rumah':
            # Pagi hari tong sampah masih kosong
            house_waste[node] = 0 
            
            # Tentukan beban sampah harian rumah ini berdasarkan bobot probabilitas
            total_harian = random.choices(berat_opsi, weights=berat_bobot, k=1)[0]
            
            if total_harian > 0:
                # Tentukan frekuensi buang (tidak boleh melebihi jumlah kg sampah)
                frekuensi = random.choices(freq_opsi, weights=freq_bobot, k=1)[0]
                frekuensi = min(frekuensi, total_harian)
                
                # Memecah total muatan menjadi beberapa kantong sampah sesuai frekuensi
                potongan_sampah = []
                sisa = total_harian
                for i in range(frekuensi - 1):
                    # Pembagian acak tapi memastikan masih ada sisa untuk jatah buang berikutnya
                    buang = random.randint(1, sisa - (frekuensi - 1 - i))
                    potongan_sampah.append(buang)
                    sisa -= buang
                potongan_sampah.append(sisa)
                
                # Penjadwalan Waktu Pembuangan (Bias Pagi Hari)
                for muatan in potongan_sampah:
                    # 75% kemungkinan orang buang sampah pagi (06:00 - 09:00 / Menit 360-540)
                    # 25% kemungkinan buang agak siang/sore (09:01 - 16:00 / Menit 541-960)
                    if random.random() <= 0.75:
                        waktu_buang = random.randint(360, 540)
                    else:
                        waktu_buang = random.randint(541, 960)
                        
                    if waktu_buang not in waste_events:
                        waste_events[waktu_buang] = []
                    waste_events[waktu_buang].append((node, muatan))

            total_sampah += total_harian
            jumlah_rumah += 1
            graf.nodes[node]['sampah_harian_total'] = total_harian

    total_initial_waste = total_sampah
    return jumlah_rumah, total_sampah

# ==============================================================================
# EKSEKUSI PEMBUATAN JADWAL
# ==============================================================================
rumah_diupdate, total_sampah_desa = generate_jadwal_sampah(G)

print("="*65)
print("     STATUS GENERASI JADWAL SAMPAH DINAMIS (REALISTIC MODE)")
print("="*65)
print(f"-> Diinisialisasi di                        : {rumah_diupdate} rumah penduduk.")
print(f"-> Prediksi total sampah terbuang hari ini  : {total_sampah_desa} kg.")
print(f"-> Total Event (Warga keluar rumah)         : {sum(len(v) for v in waste_events.values())} kejadian.")
print("="*65)

# Analisis Distribusi (Untuk membuktikan ke dosen bahwa probabilitasnya bekerja)
print("\n[VALIDASI] Rentang Waktu Warga Membuang Sampah:")
event_pagi = sum(1 for waktu in waste_events.keys() if 360 <= waktu <= 540)
event_siang = sum(1 for waktu in waste_events.keys() if waktu > 540)
print(f" - Pagi Hari (06:00-09:00) : {event_pagi} menit sibuk (Dominan)")
print(f" - Siang/Sore (09:01-16:00): {event_siang} menit sibuk")

print("\n[SAMPEL] Timeline 5 Pembuangan Paling Awal (Subuh/Pagi):")
jadwal_terurut = sorted(waste_events.keys())
tampil_max = min(5, len(jadwal_terurut))

for i in range(tampil_max):
    menit = jadwal_terurut[i]
    jam_format = f"{menit // 60:02d}:{menit % 60:02d}"
    kejadian = waste_events[menit]
    for aksi in kejadian:
         print(f" - Pukul {jam_format} | {aksi[0]} membuang {aksi[1]} kg sampah ke jalan.")
```

```python
# ==============================================================================
# CELL 3: INISIALISASI ARMADA (SISTEM SHIFT DINAMIS & TRACKING CSV)
# ==============================================================================
class Kendaraan:
    def __init__(self, id_kendaraan, tipe, kapasitas_maks, waktu_mulai, waktu_selesai, posisi_awal):
        self.id = id_kendaraan
        self.tipe = tipe 
        self.kapasitas_maks = kapasitas_maks
        self.muatan = 0
        
        self.waktu_mulai = waktu_mulai
        self.waktu_selesai = waktu_selesai
        
        # Waktu ketersediaan aktual (berubah-ubah seiring kendaraan berjalan/loading)
        self.waktu_tersedia = waktu_mulai 
        
        self.posisi = posisi_awal
        self.total_jarak = 0
        self.total_waktu_kerja = 0
        
        # Variabel pelacak statistik dan rute untuk laporan CSV & Visualisasi
        self.rumah_dikunjungi = 0
        self.kali_buang_tps = 0
        self.kali_transfer = 0
        self.rute_nodes = [posisi_awal] # Menyimpan urutan node yang dilewati
        self.log_terstruktur = []       # Menyimpan data detail untuk CSV

    def catat_log_detail(self, menit, aksi, lokasi, muatan, jarak, pesan):
        jam = menit // 60
        menit_sisa = menit % 60
        
        # Simpan dalam bentuk dictionary agar mudah di-convert ke CSV Pandas
        self.log_terstruktur.append({
            "Waktu": f"{jam:02d}:{menit_sisa:02d}",
            "ID_Armada": self.id,
            "Tipe": self.tipe,
            "Aksi": aksi,
            "Lokasi": lokasi,
            "Muatan_kg": muatan,
            "Jarak_Tempuh": jarak,
            "Total_Jarak_Sementara": self.total_jarak,
            "Pesan_Log": pesan
        })
        
        # Jika pindah lokasi, tambahkan ke jejak rute
        if self.rute_nodes[-1] != lokasi:
            self.rute_nodes.append(lokasi)

# ==============================================================================
# 1. SETUP ARMADA TRUK (Sistem Staggered Shift)
# Kapasitas 200kg. Total Jendela Operasi: 08:00 (480) - 17:00 (1020)
# ==============================================================================
armada_truk = [
    # Shift Awal: 08:00 - 14:00 (Menit 480 - 840)
    Kendaraan("Truk_1", "Truk", 200, 480, 840, "TPS_1"),
    # Shift Tengah: 10:00 - 16:00 (Menit 600 - 960)
    Kendaraan("Truk_2", "Truk", 200, 600, 960, "TPS_2"),
    # Shift Akhir: 11:00 - 17:00 (Menit 660 - 1020)
    Kendaraan("Truk_3", "Truk", 200, 660, 1020, "TPS_3")
]

# ==============================================================================
# 2. SETUP ARMADA GEROBAK (Sistem Overlapping Shift)
# Kapasitas 15kg. Total Jendela Operasi: 06:00 (360) - 15:00 (900)
# ==============================================================================
armada_gerobak = [
    # --- SHIFT PAGI (06:00 - 12:00 | Menit 360 - 720) ---
    Kendaraan("Gerobak_1", "Gerobak", 15, 360, 720, "TPS_1"),
    Kendaraan("Gerobak_2", "Gerobak", 15, 360, 720, "TPS_1"),
    Kendaraan("Gerobak_3", "Gerobak", 15, 360, 720, "TPS_2"),
    Kendaraan("Gerobak_4", "Gerobak", 15, 360, 720, "TPS_3"),

    # --- SHIFT SIANG (09:00 - 15:00 | Menit 540 - 900) ---
    Kendaraan("Gerobak_5", "Gerobak", 15, 540, 900, "TPS_2"),
    Kendaraan("Gerobak_6", "Gerobak", 15, 540, 900, "TPS_2"),
    Kendaraan("Gerobak_7", "Gerobak", 15, 540, 900, "TPS_3")
]

semua_armada = armada_gerobak + armada_truk

# Output Pengecekan
print("="*65)
print("           STATUS ARMADA SIAP OPERASI (SHIFT MODE)")
print("="*65)
print("[TRUK]    : 3 Unit terdaftar dengan Staggered Shift:")
for t in armada_truk:
    print(f"            - {t.id} | {t.waktu_mulai//60:02d}:00 - {t.waktu_selesai//60:02d}:00")

print("\n[GEROBAK] : 7 Unit terdaftar dengan Overlapping Shift:")
pagi = sum(1 for g in armada_gerobak if g.waktu_mulai == 360)
siang = sum(1 for g in armada_gerobak if g.waktu_mulai == 540)
print(f"            - {pagi} Unit Shift Pagi  (06:00 - 12:00)")
print(f"            - {siang} Unit Shift Siang (09:00 - 15:00)")
print("="*65)
print("-> Cell 3 Berhasil Diupdate: Sistem Logging Terstruktur & Rute Siap.")
```

```python
import math

print("="*75)
print("       [ENGINE START] MESIN SIMULASI CVRPTW DINAMIS (06:00 - 17:00)")
print("="*75)

# Fungsi Utilitas: Mengambil jarak dari pre-computation Dijkstra (Cell 1)
def get_jarak(node_a, node_b):
    if node_a == node_b: return 0
    return distance_matrix[node_a][node_b]

# ==============================================================================
# TIME-TICK LOOP: Simulasi berjalan menit per menit dari 06:00(360) ke 17:00(1020)
# ==============================================================================
for menit_sekarang in range(360, 1021):
    
    # 1. TRIGGER EVENT DINAMIS: Cek apakah ada warga yang membuang sampah di menit ini
    if menit_sekarang in waste_events:
        for node, muatan in waste_events[menit_sekarang]:
            house_waste[node] += muatan

    # 2. FILTER ARMADA: Cari armada yang sedang dalam jam shift-nya & sedang tidak sibuk
    armada_aktif = [k for k in semua_armada 
                    if k.waktu_mulai <= menit_sekarang < k.waktu_selesai 
                    and k.waktu_tersedia <= menit_sekarang]
    
    for k in armada_aktif:
        
        # ----------------------------------------------------------------------
        # KONDISI A: KENDARAAN PENUH ATAU SHIFT HAMPIR HABIS (< 60 Menit)
        # ----------------------------------------------------------------------
        if k.muatan >= k.kapasitas_maks or (k.waktu_selesai - menit_sekarang < 60 and k.muatan > 0):
            
            target_buang = None
            jarak_tempuh = float('inf')
            is_transfer_truk = False
            truk_target = None
            
            # Cari TPS terdekat sebagai target default
            tps_terdekat = min(TPS_NODES, key=lambda x: get_jarak(k.posisi, x))
            jarak_tps = get_jarak(k.posisi, tps_terdekat)
            
            # [LOGIKA RENDEZVOUS]: Jika yang penuh Gerobak, cek Truk yang sedang melintas!
            if k.tipe == "Gerobak":
                truk_tersedia = [t for t in armada_truk 
                                 if t.waktu_mulai <= menit_sekarang < t.waktu_selesai 
                                 and t.waktu_tersedia <= menit_sekarang # Truk juga harus sedang tidak sibuk
                                 and (t.kapasitas_maks - t.muatan) >= k.muatan]
                
                if truk_tersedia:
                    # Cari Truk yang posisinya paling dekat dengan Gerobak
                    truk_pilihan = min(truk_tersedia, key=lambda t: get_jarak(k.posisi, t.posisi))
                    jarak_truk = get_jarak(k.posisi, truk_pilihan.posisi)
                    
                    # Cerdas Memilih: Jika ke Truk lebih dekat daripada pulang ke TPS, kejar Truk!
                    if jarak_truk < jarak_tps:
                        target_buang = truk_pilihan.posisi
                        jarak_tempuh = jarak_truk
                        is_transfer_truk = True
                        truk_target = truk_pilihan
            
            if not is_transfer_truk:
                target_buang = tps_terdekat
                jarak_tempuh = jarak_tps
                
            sampah_dibawa = k.muatan
            
            # Kalkulasi Kecepatan & Waktu Loading Sesuai Ketentuan (Poin C & D)
            if k.tipe == "Gerobak":
                waktu_jalan = jarak_tempuh * 3
                waktu_load = sampah_dibawa * 2
            else: # Truk
                waktu_jalan = math.ceil(jarak_tempuh / 5) * 3
                waktu_load = math.ceil(sampah_dibawa / 10) * 2
                
            waktu_total = waktu_jalan + waktu_load
            
            # Eksekusi Keberangkatan jika sisa shift masih cukup
            if menit_sekarang + waktu_total <= k.waktu_selesai:
                k.waktu_tersedia = menit_sekarang + waktu_total
                k.total_jarak += jarak_tempuh
                k.total_waktu_kerja += waktu_total
                k.posisi = target_buang
                k.muatan = 0
                
                if is_transfer_truk:
                    # Proses sinkronisasi antar armada
                    k.kali_transfer += 1
                    truk_target.muatan += sampah_dibawa
                    # Truk harus menunggu di tempat selama gerobak jalan dan memindahkan sampah
                    truk_target.waktu_tersedia = menit_sekarang + waktu_jalan + waktu_load 
                    truk_target.total_waktu_kerja += waktu_load
                    
                    # Catat untuk laporan CSV
                    k.catat_log_detail(menit_sekarang, "Transfer ke Truk", target_buang, sampah_dibawa, jarak_tempuh, f"Rendezvous dengan {truk_target.id}")
                    truk_target.catat_log_detail(menit_sekarang, "Terima Transfer", target_buang, sampah_dibawa, 0, f"Menerima muatan dari {k.id}")
                else:
                    k.kali_buang_tps += 1
                    G.nodes[target_buang]['kapasitas_sisa'] -= sampah_dibawa
                    G.nodes[target_buang]['terisi'] += sampah_dibawa
                    
                    k.catat_log_detail(menit_sekarang, "Buang ke TPS", target_buang, sampah_dibawa, jarak_tempuh, f"Bongkar muatan di {target_buang}")
            else:
                k.waktu_tersedia = k.waktu_selesai # Pulang kandang (Shift habis)
                
        # ----------------------------------------------------------------------
        # KONDISI B: KENDARAAN KOSONG / ADA RUANG & MENCARI SAMPAH (GREEDY SEARCH)
        # ----------------------------------------------------------------------
        else:
            # Pindai seluruh rumah yang ada tumpukan sampahnya saat ini
            rumah_kotor = [n for n, w in house_waste.items() if w > 0]
            
            # Filter Akses: Truk tidak bisa memungut sampah dari Gang
            if k.tipe == "Truk":
                rumah_kotor = [n for n in rumah_kotor if G.nodes[n]['akses'] == "Semua"]
                
            if rumah_kotor:
                # Algoritma Greedy: Cari node kotor dengan jarak terpendek dari posisi sekarang
                target = min(rumah_kotor, key=lambda x: get_jarak(k.posisi, x))
                jarak_tempuh = get_jarak(k.posisi, target)
                
                sampah_tersedia = house_waste[target]
                sisa_kapasitas = k.kapasitas_maks - k.muatan
                sampah_diambil = min(sampah_tersedia, sisa_kapasitas)
                
                # Kalkulasi Waktu
                if k.tipe == "Gerobak":
                    waktu_jalan = jarak_tempuh * 3
                    waktu_load = sampah_diambil * 2
                else:
                    waktu_jalan = math.ceil(jarak_tempuh / 5) * 3
                    waktu_load = math.ceil(sampah_diambil / 10) * 2
                    
                waktu_total = waktu_jalan + waktu_load
                
                # Eksekusi Pemungutan jika sisa shift masih cukup
                if menit_sekarang + waktu_total <= k.waktu_selesai:
                    k.waktu_tersedia = menit_sekarang + waktu_total
                    k.total_jarak += jarak_tempuh
                    k.total_waktu_kerja += waktu_total
                    k.posisi = target
                    k.muatan += sampah_diambil
                    k.rumah_dikunjungi += 1
                    
                    house_waste[target] -= sampah_diambil
                    
                    k.catat_log_detail(menit_sekarang, "Ambil Sampah", target, sampah_diambil, jarak_tempuh, f"Sisa kapasitas bak: {k.kapasitas_maks - k.muatan}kg")
                else:
                    k.waktu_tersedia += 15 # Waktu mepet shift habis, biarkan istirahat tunggu shift selesai
            else:
                # Desa sedang bersih, kendaraan menunggu (idle) selama 15 menit
                k.waktu_tersedia += 15

print("-> Seluruh shift operasi selesai (17:00).")
print("-> Data spasial & temporal berhasil direkam ke dalam struktur objek armada.")
print("="*75)
```
