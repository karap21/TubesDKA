```python
import random

# Inisialisasi variabel pelacak sampah desa
house_waste = {}        
total_initial_waste = 0 
waste_events = {}       

# Fungsi membuat jadwal distribusi pembuangan sampah warga
def generate_jadwal_sampah(graf):
    global total_initial_waste, house_waste, waste_events
    total_sampah = 0
    jumlah_rumah = 0
    
    house_waste = {}
    waste_events = {}

    # Opsi berat sampah harian dan bobot peluang kemunculannya
    berat_opsi = [0, 1, 2, 3, 4, 5, 6, 7]
    berat_bobot = [0.05, 0.30, 0.35, 0.15, 0.08, 0.04, 0.02, 0.01] 
    
    # Opsi frekuensi pembuangan dalam sehari
    freq_opsi = [1, 2, 3]
    freq_bobot = [0.60, 0.35, 0.05] 

    for node, data in graf.nodes(data=True):
        if data.get('tipe') == 'Rumah':
            house_waste[node] = 0 
            
            # Memilih total berat sampah harian rumah berdasarkan bobot
            total_harian = random.choices(berat_opsi, weights=berat_bobot, k=1)[0]
            
            if total_harian > 0:
                # Memilih frekuensi buang sampah dan disesuaikan dengan total berat
                frekuensi = random.choices(freq_opsi, weights=freq_bobot, k=1)[0]
                frekuensi = min(frekuensi, total_harian)
                
                # Memecah akumulasi sampah harian menjadi beberapa kantong
                potongan_sampah = []
                sisa = total_harian
                for i in range(frekuensi - 1):
                    buang = random.randint(1, sisa - (frekuensi - 1 - i))
                    potongan_sampah.append(buang)
                    sisa -= buang
                potongan_sampah.append(sisa)
                
                # Penjadwalan waktu pembuangan ke dalam timeline menit
                for muatan in potongan_sampah:
                    # Probabilitas 75% di pagi hari (06:00 - 09:00) dan 25% di siang/sore (09:01 - 16:00)
                    if random.random() <= 0.75:
                        waktu_buang = random.randint(360, 540)
                    else:
                        waktu_buang = random.randint(541, 960)
                        
                    if waktu_buang not in waste_events:
                        waste_events[waktu_buang] = []
                    waste_events[waktu_buang].append((node, muatan))

            total_sampah += total_harian
            jumlah_rumah += 1
            graf.nodes[node]['sampah_harian_total'] = total_harian

    total_initial_waste = total_sampah
    return jumlah_rumah, total_sampah

# Eksekusi fungsi penjadwalan
rumah_diupdate, total_sampah_desa = generate_jadwal_sampah(G)

print("-> Status awal distribusi sampah dinamis berhasil dibuat.")
print(f"-> Jumlah rumah terdaftar: {rumah_diupdate} node.")
print(f"-> Total tumpukan sampah harian: {total_sampah_desa} kg.")
print(f"-> Jumlah log kejadian warga membuang sampah: {sum(len(v) for v in waste_events.values())} data.\n")

print("Sampel urutan waktu pembuangan awal:")
jadwal_terurut = sorted(waste_events.keys())
tampil_max = min(5, len(jadwal_terurut))

for i in range(tampil_max):
    menit = jadwal_terurut[i]
    jam_format = f"{menit // 60:02d}:{menit % 60:02d}"
    kejadian = waste_events[menit]
    for aksi in kejadian:
         print(f" - Jam {jam_format} | {aksi[0]} mengeluarkan sampah seberat {aksi[1]} kg.")
```

```python
class Kendaraan:
    def __init__(self, id_kendaraan, tipe, kapasitas_maks, waktu_mulai, waktu_selesai, posisi_awal):
        self.id = id_kendaraan
        self.tipe = tipe 
        self.kapasitas_maks = kapasitas_maks
        self.muatan = 0
        self.waktu_mulai = waktu_mulai
        self.waktu_selesai = waktu_selesai
        
        # Waktu operasional berjalan yang berubah dinamis saat simulasi
        self.waktu_tersedia = waktu_mulai 
        self.posisi = posisi_awal
        self.total_jarak = 0
        self.total_waktu_kerja = 0
        
        # Variabel untuk pencatatan rute dan ekspor ke dataframe pandas
        self.rumah_dikunjungi = 0
        self.kali_buang_tps = 0
        self.kali_transfer = 0
        self.rute_nodes = [posisi_awal] 
        self.log_terstruktur = []       

    def catat_log_detail(self, menit, aksi, lokasi, muatan, jarak, pesan):
        jam = menit // 60
        menit_sisa = menit % 60
        
        # Format log dictionary untuk konversi ke file CSV
        self.log_terstruktur.append({
            "Waktu": f"{jam:02d}:{menit_sisa:02d}",
            "ID_Armada": self.id,
            "Tipe": self.tipe,
            "Aksi": aksi,
            "Lokasi": lokasi,
            "Muatan_kg": muatan,
            "Jarak_Tempuh": jarak,
            "Total_Jarak_Sementara": self.total_jarak,
            "Pesan_Log": pesan
        })
        
        if self.rute_nodes[-1] != lokasi:
            self.rute_nodes.append(lokasi)

# Inisialisasi armada truk berdasarkan pembagian giliran kerja (Shift)
armada_truk = [
    Kendaraan("Truk_1", "Truk", 200, 480, 840, "TPS_1"),  # Shift 1 (08:00 - 14:00)
    Kendaraan("Truk_2", "Truk", 200, 600, 960, "TPS_2"),  # Shift 2 (10:00 - 16:00)
    Kendaraan("Truk_3", "Truk", 200, 660, 1020, "TPS_3")  # Shift 3 (11:00 - 17:00)
]

# Inisialisasi armada gerobak berdasarkan pembagian giliran kerja (Shift)
armada_gerobak = [
    # Shift Pagi (06:00 - 12:00)
    Kendaraan("Gerobak_1", "Gerobak", 15, 360, 720, "TPS_1"),
    Kendaraan("Gerobak_2", "Gerobak", 15, 360, 720, "TPS_1"),
    Kendaraan("Gerobak_3", "Gerobak", 15, 360, 720, "TPS_2"),
    Kendaraan("Gerobak_4", "Gerobak", 15, 360, 720, "TPS_3"),

    # Shift Siang (09:00 - 15:00)
    Kendaraan("Gerobak_5", "Gerobak", 15, 540, 900, "TPS_2"),
    Kendaraan("Gerobak_6", "Gerobak", 15, 540, 900, "TPS_2"),
    Kendaraan("Gerobak_7", "Gerobak", 15, 540, 900, "TPS_3")
]

semua_armada = armada_gerobak + armada_truk

print("-> Setup objek data armada kendaraan berhasil dikonfigurasi.")
print(f"-> Truk terdaftar    : {len(armada_truk)} unit.")
print(f"-> Gerobak terdaftar : {len(armada_gerobak)} unit.")
```

```python
import math

# Menghitung jarak terpendek antar dua node berdasarkan matriks Dijkstra
def get_jarak(node_a, node_b):
    if node_a == node_b: return 0
    return distance_matrix[node_a][node_b]

# Loop utama simulasi berjalan menit demi menit (06:00 hingga 17:00)
for menit_sekarang in range(360, 1021):
    
    # Update volume sampah di rumah jika ada warga yang membuang sampah di menit ini
    if menit_sekarang in waste_events:
        for node, muatan in waste_events[menit_sekarang]:
            house_waste[node] += muatan

    # Filter kendaraan yang masuk jadwal shift dan sedang posisi idle
    armada_aktif = [k for k in semua_armada 
                    if k.waktu_mulai <= menit_sekarang < k.waktu_selesai 
                    and k.waktu_tersedia <= menit_sekarang]
    
    for k in armada_aktif:
        tujuan_akhir = None
        is_pulang_ke_tps = False
        is_transfer_truk = False
        truk_target = None
        
        # 1. CEK STATUS: Apakah penuh atau mau habis waktu shift?
        if k.muatan >= k.kapasitas_maks or (k.waktu_selesai - menit_sekarang < 60 and k.muatan > 0):
            # Target default adalah TPS terdekat
            tujuan_akhir = min(TPS_NODES, key=lambda x: get_jarak(k.posisi, x))
            jarak_tps = get_jarak(k.posisi, tujuan_akhir)
            is_pulang_ke_tps = True
            
            # Logika khusus Gerobak: Cek apakah ada truk terdekat untuk transfer
            if k.tipe == "Gerobak":
                truk_tersedia = [t for t in armada_truk 
                                 if t.waktu_mulai <= menit_sekarang < t.waktu_selesai 
                                 and t.waktu_tersedia <= menit_sekarang 
                                 and (t.kapasitas_maks - t.muatan) >= k.muatan]
                
                if truk_tersedia:
                    truk_pilihan = min(truk_tersedia, key=lambda t: get_jarak(k.posisi, t.posisi))
                    jarak_truk = get_jarak(k.posisi, truk_pilihan.posisi)
                    
                    if jarak_truk < jarak_tps:
                        tujuan_akhir = truk_pilihan.posisi
                        is_transfer_truk = True
                        truk_target = truk_pilihan
                        is_pulang_ke_tps = False # Batal ke TPS, ganti kejar Truk
        else:
            # Targetnya adalah mencari rumah kotor terdekat
            rumah_kotor = [n for n, w in house_waste.items() if w > 0]
            if k.tipe == "Truk":
                rumah_kotor = [n for n in rumah_kotor if G.nodes[n]['akses'] == "Semua"]
                
            if rumah_kotor:
                tujuan_akhir = min(rumah_kotor, key=lambda x: get_jarak(k.posisi, x))
            else:
                k.waktu_tersedia += 15
                continue # Skip ke kendaraan lain jika tidak ada sampah di seluruh desa

        # Jika entah kenapa posisi sudah di tujuan akhir (misal nge-bug), selesaikan tugasnya
        if k.posisi == tujuan_akhir:
             if is_pulang_ke_tps:
                 G.nodes[k.posisi]['kapasitas_sisa'] -= k.muatan
                 G.nodes[k.posisi]['terisi'] += k.muatan
                 k.catat_log_detail(menit_sekarang, "Buang ke TPS", k.posisi, k.muatan, 0, f"Bongkar muatan di {k.posisi}")
                 k.kali_buang_tps += 1
                 k.muatan = 0
             elif is_transfer_truk:
                 k.kali_transfer += 1
                 sampah_dibawa = k.muatan
                 truk_target.muatan += sampah_dibawa
                 truk_target.waktu_tersedia = menit_sekarang + (sampah_dibawa * 2) # Waktu tunggu loading
                 truk_target.total_waktu_kerja += (sampah_dibawa * 2)
                 
                 k.catat_log_detail(menit_sekarang, "Transfer ke Truk", k.posisi, sampah_dibawa, 0, f"Rendezvous dengan {truk_target.id}")
                 truk_target.catat_log_detail(menit_sekarang, "Terima Transfer", k.posisi, sampah_dibawa, 0, f"Menerima muatan dari {k.id}")
                 k.muatan = 0
             
             k.waktu_tersedia += 5 # Istirahat sejenak setelah tugas selesai
             continue

        # 2. EKSTRAK JALUR LANGKAH DEMI LANGKAH
        jalur_lengkap = nx.dijkstra_path(G, k.posisi, tujuan_akhir, weight='weight')
        
        # 3. BERGERAK HANYA 1 LANGKAH
        node_depan = jalur_lengkap[1] 
        jarak_langkah = G[k.posisi][node_depan]['weight']
        
        if k.tipe == "Gerobak":
            waktu_jalan = jarak_langkah * 3
        else:
            waktu_jalan = math.ceil(jarak_langkah / 5) * 3
            
        waktu_load = 0
        sampah_diambil = 0
        aksi_dilakukan = "Patroli Lewat"
        pesan_log = "Melewati persimpangan"
        
        # 4. CEK KONDISI NODE DEPAN (Intercept Sampah jika ada)
        # Boleh ambil sampah JIKA BUKAN sedang dalam misi pulang/transfer, dan ada sampah di depan
        if not is_pulang_ke_tps and not is_transfer_truk and node_depan in house_waste and house_waste[node_depan] > 0:
            sampah_tersedia = house_waste[node_depan]
            sisa_kapasitas = k.kapasitas_maks - k.muatan
            sampah_diambil = min(sampah_tersedia, sisa_kapasitas)
            
            if k.tipe == "Gerobak":
                waktu_load = sampah_diambil * 2
            else:
                waktu_load = math.ceil(sampah_diambil / 10) * 2
                
            house_waste[node_depan] -= sampah_diambil
            k.muatan += sampah_diambil
            k.rumah_dikunjungi += 1
            
            aksi_dilakukan = "Ambil Sampah"
            pesan_log = f"Intercept sampah! Sisa kapasitas: {k.kapasitas_maks - k.muatan}kg"
            
        waktu_total = waktu_jalan + waktu_load
        
        # 5. UPDATE POSISI DAN LOG
        if menit_sekarang + waktu_total <= k.waktu_selesai:
            k.waktu_tersedia = menit_sekarang + waktu_total
            k.total_jarak += jarak_langkah
            k.total_waktu_kerja += waktu_total
            k.posisi = node_depan 
            
            # Jika langkah ini membuatnya tepat tiba di tujuan akhir (TPS atau Truk)
            if node_depan == tujuan_akhir:
                if is_pulang_ke_tps:
                    G.nodes[node_depan]['kapasitas_sisa'] -= k.muatan
                    G.nodes[node_depan]['terisi'] += k.muatan
                    k.catat_log_detail(menit_sekarang, "Buang ke TPS", node_depan, k.muatan, jarak_langkah, f"Bongkar muatan di {node_depan}")
                    k.kali_buang_tps += 1
                    k.muatan = 0
                elif is_transfer_truk:
                    k.kali_transfer += 1
                    sampah_dibawa = k.muatan
                    truk_target.muatan += sampah_dibawa
                    
                    # Sinkronisasi waktu truk menunggu gerobak bongkar muatan
                    truk_target.waktu_tersedia = menit_sekarang + waktu_jalan + (sampah_dibawa * 2) 
                    truk_target.total_waktu_kerja += (sampah_dibawa * 2)
                    
                    k.catat_log_detail(menit_sekarang, "Transfer ke Truk", node_depan, sampah_dibawa, jarak_langkah, f"Rendezvous sukses dengan {truk_target.id}")
                    truk_target.catat_log_detail(menit_sekarang, "Terima Transfer", node_depan, sampah_dibawa, 0, f"Menerima muatan dari {k.id}")
                    k.muatan = 0
                else:
                     k.catat_log_detail(menit_sekarang, aksi_dilakukan, node_depan, sampah_diambil, jarak_langkah, pesan_log)
            else:
                # Jika belum sampai di tujuan akhir, catat aksinya (entah cuma lewat atau ambil sampah)
                k.catat_log_detail(menit_sekarang, aksi_dilakukan, node_depan, sampah_diambil, jarak_langkah, pesan_log)
                
        else:
            k.waktu_tersedia = k.waktu_selesai 

print("-> Simulasi pergerakan logistik step-by-step selesai dijalankan.")
```
